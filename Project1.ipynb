```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Desktop/ecommerce_product_dataset.csv")

```

```python
df.head()
```

```python
# Kategoriye göre toplam satış
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

# Satışa göre büyükten küçüğe sırala
category_sales = category_sales.sort_values(by='Sales', ascending=False)

print(category_sales)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=category_sales, x='Category', y='Sales', palette='viridis')

plt.title('Kategoriye Göre Toplam Satış', fontsize=16)
plt.xlabel('Kategori', fontsize=12)
plt.ylabel('Toplam Satış', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Kategoriye göre ortalama indirim
category_discount = df.groupby('Category')['Discount'].mean().reset_index()

# Büyükten küçüğe sırala
category_discount = category_discount.sort_values(by='Discount', ascending=False)

print(category_discount)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=category_discount, x='Category', y='Discount')
plt.title('Kategoriye Göre Ortalama İndirim Oranı', fontsize=16)
plt.xlabel('Kategori', fontsize=12)
plt.ylabel('Ortalama İndirim', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price'], bins=30, kde=True)

plt.title('Ürün Fiyat Dağılımı', fontsize=16)
plt.xlabel('Fiyat', fontsize=12)
plt.ylabel('Ürün Sayısı', fontsize=12)
plt.tight_layout()
plt.show()
```
